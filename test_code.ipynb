```python
from snowball import generate_path
import numpy as np

t = np.arange(0, 243, 1)/242.0

import time
s_time = time.time()
paths = generate_path(100, 0, 0.2, t, 1000)
e_time = time.time()
print(e_time - s_time)
```

```python
import matplotlib.pyplot as plt

for i in range(paths.shape[0]):
    plt.plot(t,paths[i])
    
plt.show()
```
